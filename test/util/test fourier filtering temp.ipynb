```python
import numpy as np
import matplotlib.pyplot as plt
from tqdm.notebook import tqdm
import time
import sys
sys.path.append("../../")
from src.util import noise_generator
```

```python
N_steps = 1000
N_samples = 100000
```

```python
theta_t = np.empty((N_samples, N_steps), dtype=complex)
for i in tqdm(range(N_samples)):
    theta_t[i, :] = noise_generator.generate_white_noise(N_steps)
```

```python
xs = np.arange(N_steps)
```

```python
plt.plot(xs, np.real(np.mean(theta_t, axis=0)))
plt.plot(xs, np.imag(np.mean(theta_t, axis=0)))
plt.show()
```

```python
t = 50
plt.plot(xs, np.real(np.mean(theta_t*np.expand_dims(theta_t[:, t], -1), axis=0)))
plt.plot(xs, np.imag(np.mean(theta_t*np.expand_dims(theta_t[:, t], -1), axis=0)))
plt.show()
```

```python
t = 700
plt.plot(xs, np.real(np.mean(theta_t*np.conj(np.expand_dims(theta_t[:, t], -1)), axis=0)))
plt.plot(xs, np.imag(np.mean(theta_t*np.conj(np.expand_dims(theta_t[:, t], -1)), axis=0)))
plt.show()
```

```python
theta_k = np.fft.fft(theta_t, axis=1)
```

```python
plt.plot(xs, np.real(np.mean(theta_k, axis=0)))
plt.plot(xs, np.imag(np.mean(theta_k, axis=0)))
plt.show()
```

```python
k = 50
plt.plot(xs, np.real(np.mean(theta_k*np.expand_dims(theta_k[:, k], -1), axis=0)))
plt.plot(xs, np.imag(np.mean(theta_k*np.expand_dims(theta_k[:, k], -1), axis=0)))
plt.show()
```

```python
k = 300
plt.plot(xs, np.real(np.mean(theta_k*np.conj(np.expand_dims(theta_k[:, k], -1)), axis=0)))
plt.plot(xs, np.imag(np.mean(theta_k*np.conj(np.expand_dims(theta_k[:, k], -1)), axis=0)))
plt.show()
```
